## 5. Known metadata gaps & assumptions

Document data quality issues discovered during collection here. This section feeds directly into the limitations chapter of NB05.

| Source | Gap / Issue | Impact | Assumption / workaround |
|--------|------------|--------|------------------------|
| Oficina del Peregrino | Compostela-seekers only (~60% of actual pilgrims per Xunta 2024–25 IoT study) | Systematic undercounting | Document as structural bias; consistent across years so trend analysis valid |
| Oficina del Peregrino | Motivation categories (religious / religious+other / non-religious) are self-declared AND required for Compostela | Variable reliability | Use with caution; exclude from predictive features |
| Google Trends | Relative index (0–100), not absolute volumes | Cannot compare absolute interest across countries | Normalise within-country; use only for relative change over time |
| ERA5 | Grid resolution ~31km — route corridors are narrow | Some signal smoothing | Extract multiple grid points per route; average or take nearest |
| SJPDP data | Counts visitors to welcome office, not all pilgrims starting there | ~25–30% undercounting vs. Santiago | Use as directional signal, not absolute count |
| French routes (AFCC) | Highly fragmented; eco-counters not generalised | Incomplete coverage | Triangulate: SJPDP + eco-counters where available + association publications |
