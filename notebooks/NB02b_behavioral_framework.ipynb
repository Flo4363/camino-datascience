## Section 1 — Pilgrim Sociological Profiles

Field observation, qualitative studies (CETUR 2007–2010) and our NB02 quantitative
signals converge toward three distinct pilgrim profiles that coexist on the Camino
and have evolved in relative weight over the 2004–2024 period.

---

### Profile A — The Traditional Pilgrim

**Motivation**: primarily religious — fulfillment of a vow, devotion to Saint James,
participation in a centuries-old Catholic tradition.

**Demographic**: 50+, predominantly male, European Catholic countries
(Spain, Italy, Portugal, Poland, Germany). Often a repeat pilgrim.

**Route preference**: Camino Francés (the canonical route, rich in Romanesque
heritage and religious infrastructure). Camino Portugués for Portuguese pilgrims.

**Data signal**: `pct_religious` — declining from ~85% (2004) to ~55% (2024)
in relative terms, though stable or growing in absolute numbers.

**Behavioral characteristics**:
- Plans the pilgrimage well in advance (12–24 months)
- Familiar with the albergue system and Camino infrastructure
- May complete the route in segments over multiple years
  (staged pilgrimage behavior — see Section 4)
- Strong word-of-mouth within Catholic parish networks

**Trend**: stable or slowly declining as a share of total pilgrims.
Absolute numbers grow with the overall Camino expansion.

---

### Profile B — The Secular Seeker

**Motivation**: diffuse spiritual search, personal development, life transition
(divorce, retirement, bereavement, career change). The Camino as a ritualized
walking practice for inner transformation — what anthropologist Victor Turner
calls a *liminal* experience: a threshold crossing between life phases.

**Demographic**: 35–55 years old, predominantly female, secular or non-denominational.
Growing representation from non-Catholic countries: France, US, South Korea,
Netherlands, UK, Australia.

**Route preference**: Camino Portugués Costa (short, coastal, accessible),
Camino Inglés, Camino Portugués. Avoids the most crowded sections of the Francés.

**Data signal**: the primary driver of `pct_women` (44%→53%) and
`pct_non_religious` (5%→15%) post-2012. These two indicators are proxies
for the same underlying demographic shift (r=0.93 in NB02).

**Behavioral characteristics**:
- Often a first-time pilgrim, unlikely to repeat on the same route
- Heavily influenced by media: film (*The Way*, 2010), YouTube,
  Instagram, personal blogs and podcasts
- Discovery path: media exposure → research phase (Google Trends signal) →
  booking 6–18 months later
- Prefers private accommodation over albergues collectifs
- Less physically focused than Profile C — prioritizes emotional and
  spiritual experience over performance

**Trend**: the fastest-growing profile since 2012. Primary driver of
all structural changes observed in NB02 — feminisation, secularisation,
geographic diversification toward accessible routes.

---

### Profile C — The Athletic-Cultural Pilgrim

**Motivation**: physical challenge, nature immersion, cultural heritage,
deliberate avoidance of crowds. The Camino as an endurance and
outdoor experience first, spiritual experience second.

**Demographic**: 30–50 years old, active outdoor enthusiasts, trail runners,
hikers. Growing representation from trail-running communities (UTMB ecosystem).
Strong representation in Northern European countries (Germany, Netherlands,
Belgium, Switzerland) and increasingly South Korea.

**Route preference**: Camino del Norte (coastal, rugged, relatively uncrowded),
Camino Primitivo (the oldest Camino, significant elevation gain, Asturias
mountain passes), Vía de la Plata (extreme isolation, crosses Extremadura).
These constitute the **Challenge Cluster** identified in NB02 Axis C.

**Data signal**: `trail_countries` growth (Trail SJ by UTMB) is the
strongest leading indicator for this profile. The Trail SJ acts as
a discovery mechanism — runners encounter the Camino territory
and return as pilgrims 1–2 years later.

**Behavioral characteristics**:
- Often discovers the Camino via trail running or hiking communities
- Deliberately seeks routes with low pilgrim density
- Walks in shoulder seasons (April–May, October) when the Francés
  is most crowded — amplifying Norte/Primitivo share during these periods
- Familiar with GPS navigation, less dependent on yellow arrows
- Connected to international outdoor communities (Strava, Trail platforms)

**Trend**: growing steadily in absolute terms (Norte ×4.1, Primitivo ×4.4
since 2004). Share remains stable (~6%) as overall growth absorbs the gains.

## Section 2 — Route Behavioral Clusters

Route choice is not random — it reflects a combination of pilgrim profile,
available time, physical condition, and the specific symbolic or experiential
value the pilgrim seeks. Three behavioral clusters emerge from the NB02
all-routes analysis, each attracting a distinct profile mix.

---

### Cluster 1 — Challenge & Authenticity

**Routes**: Camino del Norte, Camino Primitivo, Vía de la Plata

**What pilgrims seek**:
- Solitude and contact with nature
- Significant physical challenge (elevation, mountain crossings,
  rugged terrain)
- Escape from the "conveyor belt" effect of the Francés post-Sarria
  (the last 100km section where Compostela-seekers concentrate)
- Authenticity — fewer pilgrims, less infrastructure, more demanding

**Primary profile**: Profile C (athletic-cultural), experienced pilgrims
repeating after the Francés, Profile A (traditional) seeking a more
challenging experience.

**Saturation hypothesis**: as the Camino Francés becomes more crowded
(media coverage of overtourism, Arte 2024 report), Challenge Cluster
routes attract a growing "saturation refugee" segment. The `share_Francés`
above a threshold may negatively predict Challenge Cluster share growth
with 1–2 year lag. → Feature `francés_saturation` in NB03.

**Vía de la Plata specificity**: despite being a Challenge route,
the Vía de la Plata is the only route in structural absolute decline
(index=82 in 2024 vs 2004 base). The extreme heat of Extremadura and
Andalusia in summer, combined with sparse infrastructure over 1,000km,
may be creating a deterrence effect that overrides the challenge appeal.
The route may be losing even Profile C pilgrims to the Norte and Primitivo,
which offer comparable challenge with more tolerable climate.

---

### Cluster 2 — Accessibility & Discovery

**Routes**: Camino Portugués Costa, Camino Inglés

**What pilgrims seek**:
- Short, manageable distance (P.Costa: ~130km Porto→Vigo,
  Inglés: ~120km Ferrol→Santiago)
- Coastal landscape, maritime heritage, scenic variety
- A first Camino experience without committing to 800km
- Easier logistical access (Porto airport for P.Costa,
  ferry from UK historically for Inglés)

**Primary profile**: Profile B (secular seeker, first-time pilgrim),
time-constrained international pilgrims (US, KR, AU).

**P.Costa growth drivers (multiple, non-exclusive)**:
1. Low-cost aviation expansion to Porto (OPO) — structural access improvement
2. Social media content in French, Korean and English communities
   presenting P.Costa as the "hidden gem" alternative
3. Post-COVID preference for outdoor, coastal, less crowded experiences
4. Trail SJ internationalisation bringing new countries into awareness
   of coastal Galicia territory

**Inglés-Korea connection**: the Camino Inglés correlates strongly
with Korean search interest (r=0.861, NB02). The short distance
and historical maritime connection may resonate with Korean pilgrims
who have limited vacation time and approach the Camino as a
once-in-a-lifetime cultural experience rather than a multi-week trek.

---

### Cluster 3 — Canonical Reference

**Routes**: Camino Francés, Camino Portugués central

**What pilgrims seek**:
- The "real" Camino — culturally anchored by film, literature,
  social media, and centuries of tradition
- Dense pilgrim community — the social dimension of the Camino
  (meeting people from around the world)
- Well-developed infrastructure — clear waymarking, abundant albergues,
  established services
- The Francés is the default choice for first-time pilgrims who
  discovered the Camino via mainstream media

**All three profiles** use these routes, but the Francés increasingly
attracts Profile B (first-time, media-driven) from long-distance countries
(US, Korea, Brazil, Australia) who replicate the cultural template
of *The Way* (SJPDP→Santiago, 800km).

**Crowding paradox**: the Francés is both the most popular route
and increasingly perceived as overcrowded — particularly the
Sarria–Santiago section (the minimum 100km for a Compostela).
This perception is driving both Challenge Cluster growth
and the P.Costa surge, creating a self-reinforcing diversification dynamic.

## Section 3 — Structural Socio-Economic Drivers

Beyond individual pilgrim motivations, the 20-year growth of the Camino
is embedded in broader structural socio-economic transformations in
Western societies that create a favorable long-term environment.
These are background forces that explain the *trend*, not the *fluctuations*.

---

### Driver 1 — Post-Materialism (Inglehart, 1977/1997)

Ronald Inglehart's theory of value change argues that post-industrial
societies progressively shift from **materialist values** (economic security,
physical safety) toward **post-materialist values** (self-expression,
quality of life, meaning, authenticity).

The Camino is a near-perfect product for post-materialist demand:
- Voluntary physical hardship as a vehicle for self-discovery
- Radical simplicity (one backpack, basic accommodation)
- Deliberate disconnection from professional and digital life
- Community with strangers across cultural boundaries
- Symbolic narrative of transformation and arrival

**Modelling implication**: the long-term growth trend of the Camino
has a structural demographic and sociological component that is
largely independent of any specific media event or route development.
Any predictive model must account for this background trend
before attributing residual variance to specific features.

---

### Driver 2 — European Demographic Aging

The 50–65 age cohort — the core of Profile A and a significant
share of Profile B — is the fastest-growing demographic segment
in Western Europe. Baby Boomers reaching retirement with
disposable income, available time, and existential questions
about meaning and legacy.

This creates a **mechanical correlation** between Camino growth
and European demographic trends that is unrelated to any specific
Camino-specific driver. The plateau in diversity (div≈0.70 post-2019)
may partly reflect the maturation of this demographic cohort.

---

### Driver 3 — Democratization of Slow Travel

Low-cost aviation expansion from the early 2000s progressively
reduced the cost and complexity of accessing Camino starting points:
- Porto (OPO): expansion of Ryanair/easyJet routes from Paris,
  London, Brussels, Amsterdam → direct enabler of P.Costa growth
- Biarritz/Pau: access for Via Podiensis pilgrims
- Bilbao/San Sebastián: access for Norte and Primitivo pilgrims

**H1 hypothesis (Axis C)**: the 2016–2018 inflection of P.Costa
may be causally linked to Porto airport route expansion during
the same period. *Testable with OAG/OpenFlights data.*
If confirmed, this would be one of the strongest causal mechanisms
in the dataset — a supply-side (infrastructure) explanation
rather than a demand-side (media/motivation) one.

---

### Driver 4 — Digital Social Networks as Amplification Infrastructure

YouTube, Instagram, TikTok and dedicated Camino forums
(Gronze.com, Camino de Santiago Forum, Reddit r/CaminodeSantiago)
have created a permanent, self-sustaining content ecosystem
that continuously introduces new potential pilgrims to the Camino.

Key dynamics:
- **Discovery loop**: pilgrim walks → creates content → inspires
  future pilgrims → they walk and create content
- **Route recommendation economy**: popular YouTubers (notably
  Korean and French channels) create strong route preferences
  in their audience communities
- **Negative feedback loop**: overcrowding content about the
  Francés drives search toward alternatives
  (captured partially by Trends DE negative signal in NB02)

Our Google Trends data captures the *aggregate demand signal*
from this ecosystem but cannot distinguish content-driven
from word-of-mouth driven discovery.

---

### Driver 5 — The COVID Meaning Shift

The COVID pandemic (2020–2021) produced a documented shift in
travel preferences toward outdoor, meaningful, slow experiences.
Multiple surveys across European countries recorded increased
interest in nature tourism, pilgrimage, and long-distance walking
post-pandemic.

Combined with deferred demand (pilgrims who postponed 2020–2021
plans), this explains the exceptional 2022–2024 growth (144% of
2019 levels by 2024). The COVID shock acted as a *meaning catalyst*
— confronting populations with mortality, isolation and the
fragility of ordinary life, directly priming the existential
motivations that drive Profile B.

## Section 4 — Key Behavioral Confounds

This section documents the behavioral patterns that create
**spurious or misleading correlations** in our quantitative data.
Understanding these confounds is essential for NB03 feature design
and NB04 model interpretation.

---

### Confound 1 — Staged Pilgrimage Behavior

**What it is**: many European pilgrims — particularly French, German,
Belgian and Swiss — complete the full Camino (Le Puy→Santiago, ~1,500km)
over multiple years, walking the French section one year and resuming
at Saint-Jean-Pied-de-Port the following year for the Spanish section.

**Evidence**: documented in CETUR 2007–2010 surveys; illustrated by
popular YouTube content (*Absent du Bureau* — "1 Million de Pas";
*Philippe Prohom* — two-part series); reflected in *The Way* (2010)
which covers only the Spanish section.

**Impact on our data**: the correlation
`via_podiensis_fr_summer_temp (T) → share_Camino_Francés (T+1)`
may be partially or entirely explained by this behavior —
pilgrims walking France in year T naturally appear in Spanish
Camino Francés statistics in year T+1, *independently of weather*.

**Two coexisting behavioral sub-profiles**:
- *European sequential*: France T → Spain T+1, may vary route
  on return trips. Predominantly FR, DE, BE, CH, IT.
- *Long-distance integral*: full route in one trip (Le Puy→Santiago)
  OR Spanish section only (SJPDP→Santiago). Predominantly US, KR, BR, AU.

**NB03 action**: build `ratio_sequential` feature
(SJPDP_foot_arrivals / Santiago_total) as a proxy for the
proportion of sequential pilgrims in a given year.
Test whether climate-route correlations persist after controlling for it.

---

### Confound 2 — Muxía-Finisterre Double Counting

**What it is**: pilgrims arriving in Santiago who continue
to the Atlantic coast (Muxía and/or Finisterre) are counted
twice in Oficina statistics — once for their primary route
and once for the Muxía-Finisterre extension.

**Impact**: Muxía-Finisterre counts are not an independent
route choice signal — they are a function of total Santiago
arrivals and the proportion who choose to continue.

**NB03/NB04 action**: exclude Muxía-Finisterre from all
route share targets and diversity index calculations.
Confirmed in NB02 — `diversity_core` already excludes it.

---

### Confound 3 — COVID Sanitary vs Structural Decomposition

**What it is**: the post-COVID Francés→P.Costa shift (-8.9pp/+9.6pp)
conflates two mechanisms that cannot be separated with available data:

*Conjunctural (COVID-specific)*:
- Albergue capacity restrictions disproportionately affected
  the Francés (large dormitories) vs coastal routes (smaller venues)
- Older pilgrim profiles (46–65, 65+) — disproportionate on the
  Francés — avoided crowded routes during sanitary uncertainty

*Structural (pre-existing trend)*:
- P.Costa inflection at 2018 (PELT confirmed) — predates COVID by 2 years
- Axis B signals (trail_countries, trends_FR) already predicted P.Costa
  growth before 2020

**NB03 action**: build `sanitary_regime` binary (2020–2022 = 1)
and test interaction with route share targets.
The 2025 data will be the first clean post-sanitary observation.

---

### Confound 4 — Common Trend Inflation

**What it is**: many of our predictors (Trends FR, Trends KR,
Trail finishers) and targets (secondary route shares) are
simultaneously growing time series. This creates positive
correlations that partly reflect the common upward trend
rather than specific causal relationships.

**Impact**: the NB02 correlation matrix is almost entirely red —
nearly every growing signal correlates positively with every
growing route. The analytically valuable signals are the
exceptions: unexpected pairings, negative correlations,
and differences in lag structure.

**NB03 action**: apply first-differencing or detrending to
time series before computing feature importance.
Test models on detrended series to isolate year-to-year
variation from the background trend.

---

### Confound 5 — Holy Year 2021 Compound Event

**What it is**: the 2021 Holy Year (Año Santo Jacobeo) coincides
exactly with the post-COVID rebound. These two effects are
statistically inseparable — no control condition exists.

**Impact**: the 2021 amplification factor (×3.25 vs 2020)
and subsequent non-correction (2022 = ×2.5 above 2021)
cannot be attributed to either Holy Year effect or
post-COVID rebound independently.

**NB03 action**: treat 2021 as a separate categorical value
(`holy_year_compound = 1`) distinct from standard Holy Years
(2004, 2010). Consider sensitivity analysis: model with
and without 2020–2022 period.

## Section 6 — Implications for NB03 & NB04

This section translates the behavioral framework into concrete
constraints and design decisions for feature engineering and modelling.

---

### Feature engineering constraints (NB03)

**From Profile analysis**:
- `pct_women` and `pct_non_religious` are proxies for the same
  underlying shift (Profile B growth) — treat as alternative
  representations of the same latent variable, not independent features
- Consider building a composite `new_pilgrim_index` that captures
  both dimensions, reducing multicollinearity

**From Route cluster analysis**:
- Build `challenge_routes_share` (Norte + Primitivo + Vía de la Plata)
  as a cluster-level feature
- Build `accessibility_routes_share` (P.Costa + Inglés)
  as a cluster-level feature
- These cluster features may be more stable and interpretable
  than individual route shares for modelling

**From Confound analysis**:
- `ratio_sequential`: SJPDP_foot / Santiago_total — controls
  staged pilgrimage behavior in Axis A climate features
- `sanitary_regime`: binary 2020–2022 — controls COVID-specific
  infrastructure constraints
- `holy_year_compound`: binary 2021 only — distinct from standard HY
- Apply detrending before computing correlations in modelling phase

**From Structural drivers analysis**:
- The long-term trend component (post-materialism + demographics)
  should be modelled explicitly as a trend feature (year, year², log_year)
  rather than absorbed into residuals
- Low-cost aviation feature for P.Costa (Porto OPO route count
  if OAG data accessible) — would test H1 directly

---

### Modelling constraints (NB04)

**Model selection**:
- n=21 observations — severely limits model complexity
- Maximum 3–4 features per model to avoid overfitting
- Prefer interpretable models (linear regression, Ridge/Lasso)
  over black-box models (XGBoost, neural networks) given the
  need for behavioral interpretation in NB05
- Consider time-series cross-validation (forward walk)
  rather than random split given temporal structure

**Interpretation constraints**:
- The model predicts route shares and volumes, not human behavior
- A good R² does not validate the causal mechanism — it only
  confirms the predictive association
- All significant predictors must be triangulated against a
  behavioral hypothesis before being reported in NB05
- Features that are statistically significant but behaviorally
  implausible should be flagged and investigated, not celebrated

**Known limitations to pre-register for NB05**:
1. Compostela-only counting (~40% undercount — IoT Xunta study)
2. Staged pilgrimage confound (ratio_sequential proxy, not direct measure)
3. COVID sanitary vs structural decomposition (unresolved without 2025 data)
4. 2021 Holy Year + COVID compound (statistically inseparable)
5. Google Trends relative index only (not absolute search volumes)
6. ERA5 spatial averaging (masks within-corridor microclimates)
7. French route data absent (AFCC contact pending)
8. Nationality data 2004–2018 only (PDFs)
9. Behavioral motivation data not available at route level
10. Long-distance aviation data (Porto OPO) not yet collected
